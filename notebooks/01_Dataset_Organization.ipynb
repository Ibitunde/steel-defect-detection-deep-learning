```python
"""
================================================================================
STEEL SURFACE DEFECT DETECTION PROJECT
NOTEBOOK 1: DATASET ORGANIZATION

Author: Mayowa Oluwaseun Ibitunde
Student ID: 23096291
Programme: MSc Data Science
University: University of Hertfordshire
Supervisor: Vito Graffagnino

Purpose: Organize NEU-CLS raw dataset files into structured class folders
================================================================================
"""

# ============================================================================
# SECTION 1: MOUNT GOOGLE DRIVE
# ============================================================================

print("="*80)
print("SECTION 1: MOUNTING GOOGLE DRIVE")
print("="*80)

from google.colab import drive
drive.mount('/content/drive')

print("\n✓ Google Drive mounted successfully!")
print("="*80)
print()


# ============================================================================
# SECTION 2: IMPORT REQUIRED LIBRARIES
# ============================================================================

print("="*80)
print("SECTION 2: IMPORTING LIBRARIES")
print("="*80)

import os
import shutil
from pathlib import Path
from collections import defaultdict

print("\n✓ All libraries imported successfully!")
print("="*80)
print()


# ============================================================================
# SECTION 3: DEFINE PATHS AND CLASS MAPPINGS
# ============================================================================

print("="*80)
print("SECTION 3: SETTING UP PATHS AND CLASS DEFINITIONS")
print("="*80)

# Define paths
RAW_DATA_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/NEU-Dataset_Raw')
ORGANIZED_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/NEU-Organized')

# Define class mappings
# Each defect type has a short prefix in the filename and a full folder name
CLASS_MAPPINGS = {
    'Cr': 'crazing',           # Surface cracks
    'In': 'inclusion',         # Non-metallic inclusions
    'Pa': 'patches',           # Surface patches
    'PS': 'pitted_surface',    # Pitted surface defects
    'RS': 'rolled_in_scale',   # Scale rolled into surface (unique to hot-rolling)
    'Sc': 'scratches'          # Surface scratches
}

print(f"\nRaw data directory: {RAW_DATA_DIR}")
print(f"Organized output directory: {ORGANIZED_DIR}")
print(f"\nDefect classes to organize: {len(CLASS_MAPPINGS)}")
for prefix, folder_name in CLASS_MAPPINGS.items():
    print(f"  {prefix} → {folder_name}")

print("\n✓ Paths and class mappings defined!")
print("="*80)
print()


# ============================================================================
# SECTION 4: CREATE ORGANIZED DIRECTORY STRUCTURE
# ============================================================================

print("="*80)
print("SECTION 4: CREATING ORGANIZED DIRECTORY STRUCTURE")
print("="*80)

# Create main organized directory
ORGANIZED_DIR.mkdir(exist_ok=True)
print(f"\n✓ Created main directory: {ORGANIZED_DIR.name}/")

# Create class subdirectories
for prefix, folder_name in CLASS_MAPPINGS.items():
    class_dir = ORGANIZED_DIR / folder_name
    class_dir.mkdir(exist_ok=True)
    print(f"✓ Created subdirectory: {folder_name}/")

print("\n✓ All directories created successfully!")
print("="*80)
print()


# ============================================================================
# SECTION 5: ORGANIZE FILES INTO CLASS FOLDERS
# ============================================================================

print("="*80)
print("SECTION 5: ORGANIZING FILES INTO CLASS FOLDERS")
print("="*80)
print("\nThis may take a few minutes for 1,800+ files...")
print()

# Initialize counters
file_counts = defaultdict(int)
skipped_files = []

# Process all .bmp files
all_files = list(RAW_DATA_DIR.glob('*.bmp'))
print(f"Found {len(all_files)} .bmp files to organize")
print()

for file_path in all_files:
    filename = file_path.name
    file_organized = False

    # Check which class this file belongs to based on filename prefix
    for prefix, folder_name in CLASS_MAPPINGS.items():
        if filename.startswith(f"{prefix}_"):
            # Copy file to appropriate class folder
            dest_path = ORGANIZED_DIR / folder_name / filename
            shutil.copy2(file_path, dest_path)
            file_counts[folder_name] += 1
            file_organized = True
            break

    # Track files that don't match any known prefix
    if not file_organized:
        skipped_files.append(filename)

print("✓ File organization complete!")
print("="*80)
print()


# ============================================================================
# SECTION 6: VERIFY ORGANIZATION AND DISPLAY SUMMARY
# ============================================================================

print("="*80)
print("SECTION 6: VERIFICATION AND SUMMARY")
print("="*80)
print()

# Display organized files count
print("Files organized by class:")
print("-" * 40)
total_organized = 0

for folder_name in CLASS_MAPPINGS.values():
    count = file_counts[folder_name]
    total_organized += count
    status = "✓" if count == 300 else "⚠️"
    print(f"  {status} {folder_name:20s}: {count:4d} images")

print("-" * 40)
print(f"  TOTAL ORGANIZED: {total_organized:4d} images")
print()

# Display any skipped files
if skipped_files:
    print(f"⚠️ WARNING: {len(skipped_files)} files skipped (unknown prefix)")
    if len(skipped_files) <= 10:
        print("Skipped files:")
        for filename in skipped_files:
            print(f"  - {filename}")
else:
    print("✓ All files successfully organized!")

print()
print("="*80)
print()


# ============================================================================
# SECTION 7: VALIDATE EXPECTED DATASET SIZE
# ============================================================================

print("="*80)
print("SECTION 7: DATASET VALIDATION")
print("="*80)
print()

EXPECTED_PER_CLASS = 300
EXPECTED_TOTAL = EXPECTED_PER_CLASS * len(CLASS_MAPPINGS)

print(f"Expected images per class: {EXPECTED_PER_CLASS}")
print(f"Expected total images: {EXPECTED_TOTAL}")
print(f"Actual total organized: {total_organized}")
print()

if total_organized == EXPECTED_TOTAL:
    print("✅ SUCCESS! Dataset organization is complete and correct!")
    print(f"   All {EXPECTED_TOTAL} images properly organized into {len(CLASS_MAPPINGS)} classes")
elif total_organized < EXPECTED_TOTAL:
    print(f"⚠️ WARNING: Missing {EXPECTED_TOTAL - total_organized} images")
    print("   Some files may not have been properly named or uploaded")
else:
    print(f"⚠️ WARNING: Found {total_organized - EXPECTED_TOTAL} extra images")
    print("   There may be duplicate files or additional data")

print()
print("="*80)
print()


# ============================================================================
# SECTION 8: DISPLAY SAMPLE FILENAMES
# ============================================================================

print("="*80)
print("SECTION 8: SAMPLE FILENAMES FROM EACH CLASS")
print("="*80)
print()

for folder_name in CLASS_MAPPINGS.values():
    class_dir = ORGANIZED_DIR / folder_name
    sample_files = sorted(class_dir.glob('*.bmp'))[:3]  # Get first 3 files

    print(f"{folder_name}:")
    for file_path in sample_files:
        print(f"  - {file_path.name}")
    print()

print("="*80)
print()


# ============================================================================
# FINAL SUMMARY
# ============================================================================

print("="*80)
print("NOTEBOOK 1 COMPLETE: DATASET ORGANIZATION")
print("="*80)
print()
print("✅ Dataset successfully organized into class folders")
print(f"✅ Output location: {ORGANIZED_DIR}")
print()
print("Next steps:")
print("  1. Verify the organized dataset in Google Drive")
print("  2. Proceed to Notebook 2: Dataset Preparation (70/10/20 split)")
print()
print("="*80)
```
