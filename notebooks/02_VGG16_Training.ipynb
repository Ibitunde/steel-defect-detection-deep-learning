```python
"""
================================================================================
STEEL SURFACE DEFECT DETECTION PROJECT
NOTEBOOK 2: DATASET PREPARATION & VGG16 TRAINING

Author: Mayowa Oluwaseun Ibitunde
Student ID: 23096291
Programme: MSc Data Science
University: University of Hertfordshire
Supervisor: Vito Graffagnino

Purpose:
- Split organized dataset into 70/10/20 (Train/Val/Test)
- Train VGG16 baseline model
- Evaluate and save results
================================================================================
"""

# ============================================================================
# SECTION 1: MOUNT GOOGLE DRIVE AND SETUP
# ============================================================================

print("="*80)
print("STEEL SURFACE DEFECT DETECTION - VGG16 BASELINE TRAINING")
print("="*80)
print()

from google.colab import drive
drive.mount('/content/drive')

print("✓ Google Drive mounted!")
print()

# Check GPU availability
import torch
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
else:
    print("⚠️ WARNING: No GPU detected! Training will be slow.")
    print("   Go to Runtime → Change runtime type → Select GPU")

print("="*80)
print()


# ============================================================================
# SECTION 2: IMPORT LIBRARIES
# ============================================================================

import os
import random
import shutil
import time
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from collections import defaultdict

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from tqdm.auto import tqdm

print("✓ All libraries imported successfully!")
print()


# ============================================================================
# SECTION 3: SET RANDOM SEEDS FOR REPRODUCIBILITY
# ============================================================================

def set_seed(seed=42):
    """
    Set random seeds for reproducibility.

    Args:
        seed (int): Random seed value
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)
print("✓ Random seed set to 42 for reproducibility")
print()


# ============================================================================
# SECTION 4: DEFINE PATHS
# ============================================================================

# Dataset paths
ORGANIZED_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/NEU-Organized')
SPLIT_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/NEU-Split')
RESULTS_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/Results')

# Create directories
RESULTS_DIR.mkdir(exist_ok=True)
(RESULTS_DIR / 'models').mkdir(exist_ok=True)
(RESULTS_DIR / 'figures').mkdir(exist_ok=True)

print(f"Organized data: {ORGANIZED_DIR}")
print(f"Split data: {SPLIT_DIR}")
print(f"Results: {RESULTS_DIR}")
print()


# ============================================================================
# SECTION 5: SPLIT DATASET (70% TRAIN / 10% VAL / 20% TEST)
# ============================================================================

print("="*80)
print("SPLITTING DATASET: 70% TRAIN / 10% VAL / 20% TEST")
print("="*80)
print()

# Get class folders
class_folders = sorted([d.name for d in ORGANIZED_DIR.iterdir() if d.is_dir()])
print(f"Found {len(class_folders)} classes: {class_folders}")
print()

# Create split directories
for split in ['train', 'val', 'test']:
    for class_name in class_folders:
        (SPLIT_DIR / split / class_name).mkdir(parents=True, exist_ok=True)

# Split each class
split_summary = {'train': 0, 'val': 0, 'test': 0}

for class_name in class_folders:
    # Get all images for this class
    class_dir = ORGANIZED_DIR / class_name
    images = sorted(list(class_dir.glob('*.bmp')))

    # Shuffle for random split
    random.shuffle(images)

    # Calculate split sizes
    n_total = len(images)
    n_train = int(n_total * 0.7)  # 70%
    n_val = int(n_total * 0.1)     # 10%
    # Remaining goes to test (20%)

    # Split images
    train_images = images[:n_train]
    val_images = images[n_train:n_train + n_val]
    test_images = images[n_train + n_val:]

    # Copy files to split directories
    for img in train_images:
        shutil.copy2(img, SPLIT_DIR / 'train' / class_name / img.name)
    for img in val_images:
        shutil.copy2(img, SPLIT_DIR / 'val' / class_name / img.name)
    for img in test_images:
        shutil.copy2(img, SPLIT_DIR / 'test' / class_name / img.name)

    # Update summary
    split_summary['train'] += len(train_images)
    split_summary['val'] += len(val_images)
    split_summary['test'] += len(test_images)

    print(f"{class_name:20s}: Train={len(train_images)}, Val={len(val_images)}, Test={len(test_images)}")

print()
print("Split Summary:")
print(f"  Training:   {split_summary['train']:4d} images ({split_summary['train']/1800*100:.1f}%)")
print(f"  Validation: {split_summary['val']:4d} images ({split_summary['val']/1800*100:.1f}%)")
print(f"  Test:       {split_summary['test']:4d} images ({split_summary['test']/1800*100:.1f}%)")
print(f"  TOTAL:      {sum(split_summary.values()):4d} images")
print()
print("✓ Dataset split complete!")
print("="*80)
print()


# ============================================================================
# SECTION 6: PYTORCH DATASET CLASS
# ============================================================================

class SteelDefectDataset(Dataset):
    """
    PyTorch Dataset for NEU Steel Surface Defect Classification.

    Args:
        root_dir (str): Directory containing class folders
        transform (callable, optional): Transform to apply to images
    """

    def __init__(self, root_dir, transform=None):
        self.root_dir = Path(root_dir)
        self.transform = transform

        # Get class names and create mapping
        self.classes = sorted([d.name for d in self.root_dir.iterdir() if d.is_dir()])
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.classes)}

        # Load all image paths and labels
        self.samples = self._load_samples()

    def _load_samples(self):
        """Load all image file paths and their labels."""
        samples = []
        for class_name in self.classes:
            class_dir = self.root_dir / class_name
            class_idx = self.class_to_idx[class_name]

            for img_path in class_dir.glob('*.bmp'):
                samples.append((str(img_path), class_idx))

        return samples

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]

        # Load image and convert to RGB
        image = Image.open(img_path).convert('RGB')

        # Apply transforms if provided
        if self.transform:
            image = self.transform(image)

        return image, label

print("✓ PyTorch Dataset class defined")
print()


# ============================================================================
# SECTION 7: DATA TRANSFORMS
# ============================================================================

# Training transforms (with augmentation to prevent overfitting)
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to VGG16 input size
    transforms.RandomHorizontalFlip(p=0.5),  # Random horizontal flip
    transforms.RandomVerticalFlip(p=0.5),    # Random vertical flip
    transforms.RandomRotation(10),            # Random rotation ±10 degrees
    transforms.ColorJitter(brightness=0.2, contrast=0.2),  # Color variation
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],  # ImageNet normalization
                        std=[0.229, 0.224, 0.225])
])

# Validation/Test transforms (NO augmentation)
val_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

print("✓ Data augmentation transforms defined")
print("  Training: Resize, Flip, Rotation, ColorJitter, Normalize")
print("  Validation/Test: Resize, Normalize only")
print()


# ============================================================================
# SECTION 8: CREATE DATASETS AND DATALOADERS
# ============================================================================

# Create datasets
train_dataset = SteelDefectDataset(SPLIT_DIR / 'train', transform=train_transform)
val_dataset = SteelDefectDataset(SPLIT_DIR / 'val', transform=val_transform)
test_dataset = SteelDefectDataset(SPLIT_DIR / 'test', transform=val_transform)

# Create dataloaders
BATCH_SIZE = 32  # Number of images per batch
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                         num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                       num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False,
                        num_workers=2, pin_memory=True)

print("="*80)
print("DATASET LOADED")
print("="*80)
print(f"Training samples:   {len(train_dataset):4d}")
print(f"Validation samples: {len(val_dataset):4d}")
print(f"Test samples:       {len(test_dataset):4d}")
print(f"Number of classes:  {len(train_dataset.classes)}")
print(f"Classes: {train_dataset.classes}")
print(f"Batch size: {BATCH_SIZE}")
print("="*80)
print()


# ============================================================================
# SECTION 9: CREATE VGG16 MODEL
# ============================================================================

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load pre-trained VGG16 (trained on ImageNet)
model = models.vgg16(pretrained=True)

# Freeze feature extraction layers (we only fine-tune the classifier)
for param in model.features.parameters():
    param.requires_grad = False

# Replace classifier for our 6 defect classes
num_features = model.classifier[0].in_features  # 25088
model.classifier = nn.Sequential(
    nn.Linear(num_features, 4096),
    nn.ReLU(inplace=True),
    nn.Dropout(0.5),
    nn.Linear(4096, 4096),
    nn.ReLU(inplace=True),
    nn.Dropout(0.5),
    nn.Linear(4096, 6)  # 6 output classes
)

model = model.to(device)

print("="*80)
print("VGG16 MODEL ARCHITECTURE")
print("="*80)
print(f"Device: {device}")
print(f"Model: VGG16 (pre-trained on ImageNet)")
print(f"Feature layers: Frozen (not trainable)")
print(f"Classifier: Fine-tuned for 6 steel defect classes")
print()

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Total parameters: {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
print(f"Frozen parameters: {total_params - trainable_params:,}")
print("="*80)
print()


# ============================================================================
# SECTION 10: TRAINING SETUP
# ============================================================================

# Loss function (CrossEntropy for multi-class classification)
criterion = nn.CrossEntropyLoss()

# Optimizer (Adam - only optimize trainable parameters)
optimizer = optim.Adam(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=0.001  # Learning rate
)

# Learning rate scheduler (reduce LR when validation loss plateaus)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',      # Minimize validation loss
    factor=0.5,      # Reduce LR by half
    patience=5       # Wait 5 epochs before reducing
)

print("✓ Training setup complete")
print(f"  Loss function: CrossEntropyLoss")
print(f"  Optimizer: Adam (lr=0.001)")
print(f"  LR Scheduler: ReduceLROnPlateau (patience=5)")
print()


# ============================================================================
# SECTION 11: TRAINING FUNCTIONS
# ============================================================================

def train_one_epoch(model, train_loader, criterion, optimizer, device):
    """
    Train model for one epoch.

    Args:
        model: PyTorch model
        train_loader: Training data loader
        criterion: Loss function
        optimizer: Optimizer
        device: Device (cuda/cpu)

    Returns:
        tuple: (average_loss, accuracy)
    """
    model.train()  # Set to training mode
    running_loss = 0.0
    correct = 0
    total = 0

    pbar = tqdm(train_loader, desc='Training', leave=False)
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)

        # Zero gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass
        loss.backward()
        optimizer.step()

        # Calculate statistics
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        # Update progress bar
        pbar.set_postfix({
            'loss': f'{loss.item():.4f}',
            'acc': f'{100.*correct/total:.2f}%'
        })

    avg_loss = running_loss / len(train_loader)
    accuracy = 100. * correct / total

    return avg_loss, accuracy


def evaluate(model, data_loader, criterion, device):
    """
    Evaluate model on validation or test set.

    Args:
        model: PyTorch model
        data_loader: Data loader
        criterion: Loss function
        device: Device (cuda/cpu)

    Returns:
        tuple: (average_loss, accuracy)
    """
    model.eval()  # Set to evaluation mode
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():  # No gradient computation
        for images, labels in data_loader:
            images, labels = images.to(device), labels.to(device)

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Calculate statistics
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    avg_loss = running_loss / len(data_loader)
    accuracy = 100. * correct / total

    return avg_loss, accuracy

print("✓ Training and evaluation functions defined")
print()


# ============================================================================
# SECTION 12: TRAIN THE MODEL
# ============================================================================

print("="*80)
print("STARTING TRAINING - VGG16 BASELINE")
print("="*80)
print()

NUM_EPOCHS = 25  # 25 epochs for faster training
best_val_acc = 0.0
history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': []
}

start_time = time.time()

for epoch in range(NUM_EPOCHS):
    print(f"Epoch [{epoch+1}/{NUM_EPOCHS}]")

    # Train
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)

    # Validate
    val_loss, val_acc = evaluate(model, val_loader, criterion, device)

    # Update learning rate
    scheduler.step(val_loss)

    # Save history
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    # Print results
    print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"  Val Loss:   {val_loss:.4f} | Val Acc:   {val_acc:.2f}%")

    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_acc': val_acc,
            'val_loss': val_loss,
        }, RESULTS_DIR / 'models/vgg16_best.pth')
        print(f"  ✓ Best model saved (Val Acc: {val_acc:.2f}%)")

    print()

total_time = time.time() - start_time

print("="*80)
print("TRAINING COMPLETE!")
print("="*80)
print(f"Total time: {total_time/60:.1f} minutes")
print(f"Best validation accuracy: {best_val_acc:.2f}%")
print("="*80)
print()


# ============================================================================
# SECTION 13: PLOT TRAINING HISTORY
# ============================================================================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

epochs_range = range(1, NUM_EPOCHS + 1)

# Loss plot
ax1.plot(epochs_range, history['train_loss'], 'b-', label='Training Loss', linewidth=2)
ax1.plot(epochs_range, history['val_loss'], 'r-', label='Validation Loss', linewidth=2)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Training and Validation Loss', fontsize=14, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Accuracy plot
ax2.plot(epochs_range, history['train_acc'], 'b-', label='Training Accuracy', linewidth=2)
ax2.plot(epochs_range, history['val_acc'], 'r-', label='Validation Accuracy', linewidth=2)
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)
ax2.set_title('Training and Validation Accuracy', fontsize=14, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(RESULTS_DIR / 'figures/vgg16_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Training curves saved!")
print()


# ============================================================================
# SECTION 14: EVALUATE ON TEST SET
# ============================================================================

# Load best model
checkpoint = torch.load(RESULTS_DIR / 'models/vgg16_best.pth')
model.load_state_dict(checkpoint['model_state_dict'])

# Evaluate on test set
test_loss, test_acc = evaluate(model, test_loader, criterion, device)

print("="*80)
print("FINAL TEST RESULTS")
print("="*80)
print(f"Test Loss:     {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.2f}%")
print("="*80)
print()

# Save final results to file
results_text = f"""
VGG16 BASELINE - STEEL DEFECT DETECTION RESULTS
================================================

Training Configuration:
- Model: VGG16 (pre-trained on ImageNet)
- Dataset: NEU Surface Defect Database
- Classes: 6 (crazing, inclusion, patches, pitted_surface, rolled_in_scale, scratches)
- Training samples: {len(train_dataset)}
- Validation samples: {len(val_dataset)}
- Test samples: {len(test_dataset)}
- Epochs: {NUM_EPOCHS}
- Batch size: {BATCH_SIZE}
- Optimizer: Adam (lr=0.001)
- Data augmentation: RandomFlip, Rotation, ColorJitter

Final Results:
- Best Validation Accuracy: {best_val_acc:.2f}%
- Test Accuracy: {test_acc:.2f}%
- Test Loss: {test_loss:.4f}
- Training time: {total_time/60:.1f} minutes
"""

with open(RESULTS_DIR / 'vgg16_results.txt', 'w') as f:
    f.write(results_text)

print("✓ Results saved to vgg16_results.txt")
print()

print("="*80)
print("🎉 VGG16 BASELINE TRAINING COMPLETE! 🎉")
print("="*80)
print()
print("Files saved:")
print(f"  ✓ Model: {RESULTS_DIR}/models/vgg16_best.pth")
print(f"  ✓ Training curves: {RESULTS_DIR}/figures/vgg16_training_curves.png")
print(f"  ✓ Results summary: {RESULTS_DIR}/vgg16_results.txt")
print()
print("Next steps:")
print("  1. Review training curves and results")
print("  2. Train additional models (ResNet50, etc.)")
print("  3. Compare model performance")
print("  4. Prepare presentation slides")
print()
print("="*80)
```
