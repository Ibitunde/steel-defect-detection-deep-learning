```python
"""
================================================================================
STEEL SURFACE DEFECT DETECTION PROJECT
NOTEBOOK 8: COMPREHENSIVE MODEL EVALUATION & COMPARISON

Author: Mayowa Oluwaseun Ibitunde
Student ID: 23096291
Programme: MSc Data Science
University: University of Hertfordshire
Supervisor: Vito Graffagnino

Purpose:
- Evaluate all 6 trained models comprehensively
- Generate confusion matrices, per-class metrics
- Compare inference times, model sizes
- Create publication-ready comparison tables
================================================================================
"""

# ============================================================================
# SECTION 1: MOUNT GOOGLE DRIVE AND SETUP
# ============================================================================

print("="*80)
print("COMPREHENSIVE MODEL EVALUATION & COMPARISON")
print("="*80)
print()

from google.colab import drive
drive.mount('/content/drive')

print("✓ Google Drive mounted!")
print()

import torch
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
print("="*80)
print()


# ============================================================================
# SECTION 2: IMPORT LIBRARIES
# ============================================================================

import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from collections import defaultdict

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.metrics import (confusion_matrix, classification_report,
                            precision_recall_fscore_support, accuracy_score)

print("✓ All libraries imported successfully!")
print()


# ============================================================================
# SECTION 3: DEFINE PATHS AND SETUP
# ============================================================================

SPLIT_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/NEU-Split')
RESULTS_DIR = Path('/content/drive/MyDrive/FINAL PROJECT/Results')
MODELS_DIR = RESULTS_DIR / 'models'
FIGURES_DIR = RESULTS_DIR / 'figures'
EVALUATION_DIR = RESULTS_DIR / 'evaluation'

EVALUATION_DIR.mkdir(exist_ok=True)

print(f"Models: {MODELS_DIR}")
print(f"Evaluation output: {EVALUATION_DIR}")
print()


# ============================================================================
# SECTION 4: DATASET CLASS
# ============================================================================

class SteelDefectDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = Path(root_dir)
        self.transform = transform
        self.classes = sorted([d.name for d in self.root_dir.iterdir() if d.is_dir()])
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.classes)}
        self.samples = self._load_samples()

    def _load_samples(self):
        samples = []
        for class_name in self.classes:
            class_dir = self.root_dir / class_name
            class_idx = self.class_to_idx[class_name]
            for img_path in class_dir.glob('*.bmp'):
                samples.append((str(img_path), class_idx))
        return samples

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, label

print("✓ Dataset class defined")
print()


# ============================================================================
# SECTION 5: LOAD TEST DATA
# ============================================================================

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

test_dataset = SteelDefectDataset(SPLIT_DIR / 'test', transform=test_transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=0)

class_names = test_dataset.classes

print(f"✓ Test dataset loaded: {len(test_dataset)} images")
print(f"✓ Classes: {class_names}")
print()


# ============================================================================
# SECTION 6: DEFINE MODEL CONFIGURATIONS
# ============================================================================

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

MODEL_CONFIGS = {
    'VGG16': {
        'checkpoint': 'vgg16_best.pth',
        'architecture': 'CNN',
        'create_fn': lambda: models.vgg16(pretrained=False),
        'modify_fn': lambda m: setattr(m, 'classifier', nn.Sequential(
            nn.Linear(25088, 4096), nn.ReLU(inplace=True), nn.Dropout(0.5),
            nn.Linear(4096, 4096), nn.ReLU(inplace=True), nn.Dropout(0.5),
            nn.Linear(4096, 6)
        ))
    },
    'ResNet50': {
        'checkpoint': 'resnet50_best.pth',
        'architecture': 'CNN',
        'create_fn': lambda: models.resnet50(pretrained=False),
        'modify_fn': lambda m: setattr(m, 'fc', nn.Linear(m.fc.in_features, 6))
    },
    'EfficientNetB0': {
        'checkpoint': 'efficientnet_b0_best.pth',
        'architecture': 'CNN',
        'create_fn': lambda: models.efficientnet_b0(pretrained=False),
        'modify_fn': lambda m: setattr(m, 'classifier', nn.Sequential(
            nn.Dropout(p=0.2, inplace=True),
            nn.Linear(m.classifier[1].in_features, 6)
        ))
    },
    'MobileNetV3-Large': {
        'checkpoint': 'mobilenet_v3_large_best.pth',
        'architecture': 'CNN',
        'create_fn': lambda: models.mobilenet_v3_large(pretrained=False),
        'modify_fn': lambda m: setattr(m, 'classifier', nn.Sequential(
            nn.Linear(m.classifier[0].in_features, 1280),
            nn.Hardswish(inplace=True),
            nn.Dropout(p=0.2, inplace=True),
            nn.Linear(1280, 6)
        ))
    },
    'ViT-Base/16': {
        'checkpoint': 'vit_base_16_best.pth',
        'architecture': 'Transformer',
        'create_fn': lambda: models.vit_b_16(pretrained=False),
        'modify_fn': lambda m: setattr(m.heads, 'head', nn.Linear(m.heads.head.in_features, 6))
    },
    'Swin-Tiny': {
        'checkpoint': 'swin_t_best.pth',
        'architecture': 'Transformer',
        'create_fn': lambda: models.swin_t(pretrained=False),
        'modify_fn': lambda m: setattr(m, 'head', nn.Linear(m.head.in_features, 6))
    }
}

print("✓ Model configurations defined")
print()


# ============================================================================
# SECTION 7: EVALUATION FUNCTIONS
# ============================================================================

def load_model(model_name, config):
    """Load a trained model."""
    print(f"Loading {model_name}...", end=" ")

    # Create model
    model = config['create_fn']()
    config['modify_fn'](model)

    # Load checkpoint
    checkpoint_path = MODELS_DIR / config['checkpoint']
    checkpoint = torch.load(checkpoint_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])

    model = model.to(device)
    model.eval()

    print("✓")
    return model


def get_predictions(model, data_loader):
    """Get all predictions and true labels."""
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device)
            outputs = model(images)
            _, preds = outputs.max(1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    return np.array(all_preds), np.array(all_labels)


def measure_inference_time(model, data_loader, num_iterations=3):
    """Measure average inference time per image."""
    times = []

    with torch.no_grad():
        for _ in range(num_iterations):
            start_time = time.time()

            for images, _ in data_loader:
                images = images.to(device)
                _ = model(images)

            elapsed = time.time() - start_time
            times.append(elapsed)

    avg_time = np.mean(times)
    time_per_image = avg_time / len(data_loader.dataset)

    return time_per_image


def count_parameters(model):
    """Count model parameters."""
    return sum(p.numel() for p in model.parameters())


print("✓ Evaluation functions defined")
print()


# ============================================================================
# SECTION 8: EVALUATE ALL MODELS
# ============================================================================

print("="*80)
print("EVALUATING ALL MODELS")
print("="*80)
print()

results = {}

for model_name, config in MODEL_CONFIGS.items():
    print(f"\n{model_name}")
    print("-" * 40)

    # Load model
    model = load_model(model_name, config)

    # Get predictions
    print("  Getting predictions...", end=" ")
    preds, labels = get_predictions(model, test_loader)
    print("✓")

    # Calculate metrics
    print("  Calculating metrics...", end=" ")
    accuracy = accuracy_score(labels, preds) * 100
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted', zero_division=0)
    conf_matrix = confusion_matrix(labels, preds)
    print("✓")

    # Measure inference time
    print("  Measuring inference time...", end=" ")
    inf_time = measure_inference_time(model, test_loader)
    print("✓")

    # Count parameters
    print("  Counting parameters...", end=" ")
    params = count_parameters(model)
    print("✓")

    # Store results
    results[model_name] = {
        'accuracy': accuracy,
        'precision': precision * 100,
        'recall': recall * 100,
        'f1_score': f1 * 100,
        'inference_time_ms': inf_time * 1000,
        'parameters': params,
        'architecture': config['architecture'],
        'predictions': preds,
        'labels': labels,
        'confusion_matrix': conf_matrix
    }

    print(f"  Test Accuracy: {accuracy:.2f}%")

print("\n" + "="*80)
print("✓ ALL MODELS EVALUATED")
print("="*80)
print()


# ============================================================================
# SECTION 9: CREATE COMPARISON TABLE
# ============================================================================

print("Creating comparison table...")

# Create DataFrame
df_results = pd.DataFrame({
    'Model': list(results.keys()),
    'Architecture': [r['architecture'] for r in results.values()],
    'Accuracy (%)': [r['accuracy'] for r in results.values()],
    'Precision (%)': [r['precision'] for r in results.values()],
    'Recall (%)': [r['recall'] for r in results.values()],
    'F1-Score (%)': [r['f1_score'] for r in results.values()],
    'Parameters (M)': [r['parameters'] / 1e6 for r in results.values()],
    'Inference (ms)': [r['inference_time_ms'] for r in results.values()]
})

# Sort by accuracy
df_results = df_results.sort_values('Accuracy (%)', ascending=False)

# Save to CSV
df_results.to_csv(EVALUATION_DIR / 'model_comparison.csv', index=False)

print("✓ Comparison table saved")
print()
print(df_results.to_string(index=False))
print()


# ============================================================================
# SECTION 10: PLOT CONFUSION MATRICES
# ============================================================================

print("Creating confusion matrices...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for idx, (model_name, data) in enumerate(results.items()):
    cm = data['confusion_matrix']
    accuracy = data['accuracy']

    # Normalize confusion matrix
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Plot
    ax = axes[idx]
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names, ax=ax,
                cbar_kws={'label': 'Proportion'})
    ax.set_title(f'{model_name}\nAccuracy: {accuracy:.2f}%', fontweight='bold', fontsize=12)
    ax.set_xlabel('Predicted Label', fontsize=10)
    ax.set_ylabel('True Label', fontsize=10)
    ax.tick_params(axis='both', labelsize=8)

plt.tight_layout()
plt.savefig(EVALUATION_DIR / 'confusion_matrices_all.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Confusion matrices saved")
print()


# ============================================================================
# SECTION 11: PER-CLASS METRICS
# ============================================================================

print("Calculating per-class metrics...")

# Create per-class comparison for each model
for model_name, data in results.items():
    preds = data['predictions']
    labels = data['labels']

    # Get per-class metrics
    report = classification_report(labels, preds, target_names=class_names,
                                  output_dict=True, zero_division=0)

    # Create DataFrame
    df_class = pd.DataFrame({
        'Class': class_names,
        'Precision (%)': [report[c]['precision'] * 100 for c in class_names],
        'Recall (%)': [report[c]['recall'] * 100 for c in class_names],
        'F1-Score (%)': [report[c]['f1-score'] * 100 for c in class_names],
        'Support': [report[c]['support'] for c in class_names]
    })

    # Save
    df_class.to_csv(EVALUATION_DIR / f'{model_name.replace("/", "_")}_per_class.csv', index=False)

print("✓ Per-class metrics saved")
print()


# ============================================================================
# SECTION 12: VISUALIZATION - ACCURACY COMPARISON
# ============================================================================

print("Creating accuracy comparison chart...")

fig, ax = plt.subplots(figsize=(12, 6))

models = df_results['Model'].tolist()
accuracies = df_results['Accuracy (%)'].tolist()
colors = ['#1f77b4' if arch == 'CNN' else '#ff7f0e'
          for arch in df_results['Architecture'].tolist()]

bars = ax.barh(models, accuracies, color=colors)

# Add value labels
for bar, acc in zip(bars, accuracies):
    ax.text(bar.get_width() - 0.5, bar.get_y() + bar.get_height()/2,
            f'{acc:.2f}%', ha='right', va='center', fontweight='bold', fontsize=10)

ax.set_xlabel('Test Accuracy (%)', fontsize=12, fontweight='bold')
ax.set_title('Model Performance Comparison', fontsize=14, fontweight='bold')
ax.set_xlim(95, 101)
ax.grid(axis='x', alpha=0.3)

# Legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#1f77b4', label='CNN'),
                  Patch(facecolor='#ff7f0e', label='Transformer')]
ax.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig(EVALUATION_DIR / 'accuracy_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Accuracy comparison chart saved")
print()


# ============================================================================
# SECTION 13: EFFICIENCY ANALYSIS
# ============================================================================

print("Creating efficiency analysis...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Accuracy vs Parameters
ax1.scatter(df_results['Parameters (M)'], df_results['Accuracy (%)'],
           c=['#1f77b4' if arch == 'CNN' else '#ff7f0e'
              for arch in df_results['Architecture']],
           s=200, alpha=0.6, edgecolors='black', linewidth=2)

for idx, row in df_results.iterrows():
    ax1.annotate(row['Model'],
                (row['Parameters (M)'], row['Accuracy (%)']),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

ax1.set_xlabel('Parameters (Millions)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Test Accuracy (%)', fontsize=12, fontweight='bold')
ax1.set_title('Accuracy vs Model Size', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(97, 101)

# Accuracy vs Inference Time
ax2.scatter(df_results['Inference (ms)'], df_results['Accuracy (%)'],
           c=['#1f77b4' if arch == 'CNN' else '#ff7f0e'
              for arch in df_results['Architecture']],
           s=200, alpha=0.6, edgecolors='black', linewidth=2)

for idx, row in df_results.iterrows():
    ax2.annotate(row['Model'],
                (row['Inference (ms)'], row['Accuracy (%)']),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

ax2.set_xlabel('Inference Time (ms per image)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Test Accuracy (%)', fontsize=12, fontweight='bold')
ax2.set_title('Accuracy vs Inference Speed', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_ylim(97, 101)

plt.tight_layout()
plt.savefig(EVALUATION_DIR / 'efficiency_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Efficiency analysis saved")
print()


# ============================================================================
# SECTION 14: FINAL SUMMARY
# ============================================================================

print("="*80)
print("COMPREHENSIVE EVALUATION COMPLETE!")
print("="*80)
print()
print("Files saved to:", EVALUATION_DIR)
print()
print("Generated files:")
print("  ✓ model_comparison.csv - Complete metrics table")
print("  ✓ confusion_matrices_all.png - All 6 confusion matrices")
print("  ✓ accuracy_comparison.png - Bar chart comparison")
print("  ✓ efficiency_analysis.png - Accuracy vs Size/Speed")
print("  ✓ *_per_class.csv - Per-class metrics for each model")
print()
print("="*80)
print("🎊 ALL ANALYSIS COMPLETE! READY FOR PRESENTATION! 🎊")
print("="*80)
```
